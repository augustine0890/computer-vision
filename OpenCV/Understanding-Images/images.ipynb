```python
import numpy as np
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = Image.open("../images/beach.png")
```

```python
img
```

```python
img_mat = matplotlib.image.imread("../images/beach.png") # RGB
img_mat.shape
```

```python
r = img_mat[:,:,0]
g = img_mat[:,:,1]
b = img_mat[:,:,2]
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.imshow(r, cmap='gray'); #cmap='gray'
plt.subplot(1,3,2)
plt.imshow(g, cmap='gray');
plt.subplot(1,3,3)
plt.imshow(b, cmap='gray');
plt.show()
```
